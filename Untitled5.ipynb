5- How has the trend of adding new content evolved year by year?

```python
plt.subplot(3, 3, 5)

# Create the 'year_added' column from 'date_added'
# The .dt.year accessor extracts the year.
df['year_added'] = df['date_added'].dt.year

# Create a clean Series for plotting by dropping NaNs and converting to integer
# This avoids errors and ensures the x-axis has integer years (e.g., 2019)
valid_years = df['year_added'].dropna().astype(int)
added_by_year = valid_years.value_counts().sort_index()

added_by_year.plot(kind='line', marker='o', color='#E50914', linewidth=2)
plt.title('Content Added to Netflix by Year')
plt.xlabel('Year Added')


plt.ylabel('Number of Titles')

plt.xticks(rotation=45)

# Calculate growth
print(f"\nQ5. Content Addition Trend:")

# Moved all logic/printing inside conditional blocks
# This prevents a NameError if 'growth_rate' is not calculated.
if len(added_by_year) > 1:
    growth_rate = ((added_by_year.iloc[-1] - added_by_year.iloc[-2]) / added_by_year.iloc[-2]) * 100
    print(f" - Latest Year: {added_by_year.index[-1]} - {added_by_year.iloc[-1]} titles")
    print(f" - Growth Rate: {growth_rate:+.1f}% from previous year")
elif len(added_by_year) == 1:
    # Handle case with only one year of data
    print(f" - Latest Year: {added_by_year.index[-1]} - {added_by_year.iloc[-1]} titles")
    print(f" - Not enough data to calculate growth.")
else:
    # Handle case with no data
    print(" - No content addition data found.")
```

**Insight**: There was a massive surge in content additions starting around 2015, peaking in 2019.  Afterward, there's a noticeable decline. This suggests an aggressive content acquisition strategy leading up to 2019, followed by a slowdown, which could be attributed to market saturation or external factors like the COVID-19 pandemic impacting production pipelines. 


**Recommendation**: Netflix should investigate the cause of the post-2019 slowdown. To maintain subscriber interest and reduce churn, it's crucial to ensure a steady, predictable stream of new content. Netflix could strategize to create "mini-peak" seasons in historically slower months to keep year-round engagement high.

8 - Whch genres are more associated with TV shows vs movies ?

```python
plt.subplot(3, 3, 8)

# Get top genres for movies and TV shows separately
movies_df = df[df['type'] == 'Movie']
tv_shows_df = df[df['type'] == 'TV Show']


movie_genres = []
for genres in movies_df['listed_in'].dropna():
    genre_list = [genre.strip() for genre in genres.split(",")]
    movie_genres.extend(genre_list)


tv_genres = []
for genres in tv_shows_df['listed_in'].dropna():
    genre_list = [genre.strip() for genre in genres.split(",")]
    tv_genres.extend(genre_list)

top_movie_genres = pd.Series(movie_genres).value_counts().head(5)
top_tv_genres = pd.Series(tv_genres).value_counts().head(5)

# Create comparison plot
genres_comparison = pd.DataFrame({
    'Movies': top_movie_genres,
    'TV Shows': top_tv_genres
}).fillna(0)

genres_comparison.plot(kind='bar', ax=plt.gca()) 
plt.title('Top Genres: Movies vs TV Shows')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)

print(f"\nQ8. Genre Comparison - Movies vs TV Shows:")
print(" Movies Top Genres:")
for i, (genre, count) in enumerate(top_movie_genres.head().items(), 1):
    print(f" {i}. {genre}: {count}")

print(" TV Shows Top Genres:")
for i, (genre, count) in enumerate(top_tv_genres.head().items(), 1):
    print(f" {i}. {genre}: {count}")
```

**Insight**: While many genres overlap, there are clear distinctions. "International Movies," "Dramas," and "Documentaries" are strongly associated with the Movie format. Conversely, categories like "International TV Shows," "TV Dramas," "Crime TV Shows," and "Kids' TV" are exclusive to or dominated by the TV Show format.  This shows a strategic separation in how different types of stories are packaged.




**Recommendation**: Netflix should continue to align genre with the optimal format. For expansive, character-driven narratives, the TV show format is ideal for building long-term engagement. For self-contained stories, the movie format is more appropriate. Cross-promoting similar genres across formats—for instance, recommending a crime documentary to a viewer who enjoys crime TV shows—could also improve content discovery.

17- Is there a trend in movie durations over the years?

```python
import numpy as np


recent_movies = df[(df['type'] == 'Movie') &
                   (df['release_year'] >= 2000) &
                   (df['movie_duration_min'].notna())]


duration_by_year = recent_movies.groupby('release_year')['movie_duration_min'].mean()

# Check if there's data to plot before proceeding
if not duration_by_year.empty:
    axes[1,0].plot(duration_by_year.index, duration_by_year.values, marker='o', color='#E50914', linewidth=2)
    axes[1,0].set_title('Average Movie Duration Trend (2000-Present)')
    axes[1,0].set_xlabel('Release Year')
    axes[1,0].set_ylabel('Average Duration (minutes)')
    axes[1,0].tick_params(axis='x', rotation=45)

    # Add trend line if enough data points
    if len(duration_by_year) > 1:
        z = np.polyfit(duration_by_year.index, duration_by_year.values, 1)
        p = np.poly1d(z)
        axes[1,0].plot(duration_by_year.index, p(duration_by_year.index), "r--", alpha=0.5)

    plt.tight_layout()
    plt.show()

    print(f"\nQ17. Movie Duration Trend:")
    print(f" - Average duration in earliest year ({duration_by_year.index[0]}): {duration_by_year.iloc[0]:.1f} min")
    print(f" - Average duration in latest year ({duration_by_year.index[-1]}): {duration_by_year.iloc[-1]:.1f} min")
    
    change = duration_by_year.iloc[-1] - duration_by_year.iloc[0]
    print(f" - Change over period: {change:+.1f} minutes")
else:
    print("\nQ17. No movie duration data found for the specified period.")
    # You might want to display the empty plot with a title
    axes[1,0].set_title('Average Movie Duration Trend (2000-Present)')
    plt.tight_layout()
    plt.show()
```

**Insight**: Yes, there is a clear downward trend. The average movie duration has decreased by approximately 14.6 minutes between 2000 and 2021, dropping from around 111 minutes to 96 minutes.  This suggests a shift in filmmaking or viewer preference towards shorter, more concise feature films.


**Recommendation**: While the standard 90–110 minute feature film remains a safe bet, Netflix should embrace this trend by commissioning and acquiring more shorter-form movies (e.g., 75–90 minutes).  Marketing these as "quick watches" or "weeknight movies" could appeal to viewers with limited time and cater to modern attention spans

```python

all_genres_series = df['listed_in'].str.split(', ').explode()
genre_counts = all_genres_series.value_counts()


recent_content = df[df['release_year'] >= 2010]


top_genres_overall = genre_counts.head(5).index

genre_year_data = []
for year in sorted(recent_content['release_year'].unique()):
  
    year_data = recent_content[recent_content['release_year'] == year]
    
    
    year_genres = []
    
    for genres in year_data['listed_in'].dropna():
       
        genre_list = [genre.strip() for genre in genres.split(',')]
        year_genres.extend(genre_list)
        
    genre_counts_year = pd.Series(year_genres).value_counts()
    
    for genre in top_genres_overall:
        genre_year_data.append({
            'Year': year,
            'Genre': genre,
            'Count': genre_counts_year.get(genre, 0)
        })

genre_year_df = pd.DataFrame(genre_year_data)

# Check if DataFrame is not empty before pivoting and plotting
if not genre_year_df.empty:
    genre_pivot = genre_year_df.pivot(index='Year', columns='Genre', values='Count').fillna(0)
    genre_pivot.plot(kind='line', ax=axes[0,1], linewidth=2, marker='o')
    
    
    axes[0,1].set_title('Top Genres Distribution Over Years (2010-Present)')
    axes[0,1].set_xlabel('Release Year')
    
    
    axes[0,1].set_ylabel('Number of Titles')
    axes[0,1].legend(title='Genre')

    print(f"\nQ19. Genre Distribution Trends:")
    
 
    for genre in top_genres_overall:
        genre_trend = genre_pivot[genre]
        # Ensure there are at least two data points to compare
        if len(genre_trend.dropna()) > 1:
            change = genre_trend.iloc[-1] - genre_trend.iloc[0]
            trend = "increasing" if change > 5 else "decreasing" if change < -5 else "stable"
        else:
            trend = "stable (not enough data)"
        print(f" - {genre}: {trend} trend")
else:
    print("\nQ19. No data found for genre distribution trends in the last 10 years.")
    axes[0,1].set_title('Top Genres Distribution Over Years (2010-Present)')
```

**Insight**: The top five genres—International Movies, Dramas, Comedies, International TV Shows, and Documentaries—have all shown an increasing trend in production over the last decade (since 2010).  This indicates that Netflix's growth has been built on scaling up its core, most popular content categories rather than diversifying into radically new ones.

**Recommendation**: Netflix should continue to scale production in these proven genres. However, to avoid content fatigue, it should focus on innovation within these categories. For example, exploring sub-genres (like romantic comedies or historical dramas), experimenting with hybrid genres, and sourcing content from new international markets can keep the offerings fresh while still catering to established audience tastes.

20 - Which countries produce the most content in each genre?

```python
# Analyze which countries produce the most content in top genres


all_genres_series = df['listed_in'].str.split(', ').explode().str.strip()
genre_counts = all_genres_series.value_counts()
top_5_genres = genre_counts.head(5).index

country_genre_analysis = {}
for genre in top_5_genres:
    
    genre_content = df[df['listed_in'].str.contains(genre, na=False)]
    
   
    genre_countries = []
    
   
    for countries in genre_content['country'].dropna():
        if countries != 'Unknown':
           
            country_list = [country.strip() for country in countries.split(",")]
            genre_countries.extend(country_list)
            
    top_countries = pd.Series(genre_countries).value_counts().head(3)
    country_genre_analysis[genre] = top_countries

# Create a summary table

summary_data = []
for genre, countries in country_genre_analysis.items():
    for country, count in countries.items():
        summary_data.append({'Genre': genre, 'Country': country, 'Count': count})

summary_df = pd.DataFrame(summary_data)

# Check if the summary DataFrame is empty before proceeding
if not summary_df.empty:
    
    pivot_table = summary_df.pivot(index='Genre', columns='Country', values='Count').fillna(0)
    pivot_table.plot(kind='bar', ax=axes[1,0], width=0.8)

    
    axes[1,0].set_title('Top Producing Countries by Genre')
    axes[1,0].set_xlabel('Genre')
    axes[1,0].set_ylabel('Number of Titles')
    axes[1,0].tick_params(axis='x', rotation=45)
    axes[1,0].legend(title='Country')

    plt.tight_layout()
    plt.show()

    print(f"\nQ20. Top Countries by Genre Production:")
    for genre in top_5_genres:
        if genre in country_genre_analysis and not country_genre_analysis[genre].empty:
            top_country = country_genre_analysis[genre].index[0]
            top_count = country_genre_analysis[genre].iloc[0]
            print(f" - {genre}: {top_country} ({top_count} titles)")
        else:
            print(f" - {genre}: No country data found.")
else:
    print("\nQ20. No data available to analyze top producing countries by genre.")
    axes[1,0].set_title('Top Producing Countries by Genre')
    plt.tight_layout()
    plt.show()
    
```

**Insight**: The United States is the leader in producing "Dramas," "Comedies," and "Documentaries." India is the top producer of "International Movies," while South Korea leads in "International TV Shows."  This highlights distinct national strengths in content production.



**Recommendation**: Netflix should adopt a tailored acquisition strategy that plays to these national strengths. For high-quality dramas and comedies, partnerships with U.S. studios are key. For captivating international films, collaborations with Indian production houses should be prioritized. Similarly, for binge-worthy international series, investing in South Korean talent and studios is a proven strategy.

CONTENT LAG ANALYSIS

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure required columns are numeric and without NaNs
df_analysis = df[['release_year', 'year_added']].dropna()
df_analysis = df_analysis.astype(int)

# Calculate the content lag
df_analysis['content_lag'] = df_analysis['year_added'] - df_analysis['release_year']

# Filter for realistic lags (e.g., content added after or on the release year)
df_analysis = df_analysis[df_analysis['content_lag'] >= 0]

plt.figure(figsize=(12, 6))
sns.histplot(df_analysis['content_lag'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Content Lag (Years Between Release and Addition to Netflix)')
plt.xlabel('Years Lag')
plt.ylabel('Number of Titles')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print key statistics
print("--- Content Lag Analysis ---")
print(f"Average lag time: {df_analysis['content_lag'].mean():.1f} years")
print(f"Median lag time: {df_analysis['content_lag'].median():.1f} years")
```

RATING MATURITY LEVEL BREAKDOWN

```python
# Define maturity categories based on ratings
def assign_maturity(rating):
    if rating in ['TV-Y', 'TV-Y7', 'TV-G', 'G', 'PG', 'TV-PG']:
        return 'Kids & Family'
    elif rating in ['PG-13', 'TV-14']:
        return 'Teens'
    elif rating in ['R', 'TV-MA', 'NC-17']:
        return 'Adults'
    else:
        return 'Unrated/Other'

df['maturity_level'] = df['rating'].apply(assign_maturity)
maturity_counts = df['maturity_level'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(maturity_counts, labels=maturity_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Content Distribution by Maturity Level')
plt.ylabel('') # Hides the 'maturity_level' label on the y-axis
plt.show()
```
